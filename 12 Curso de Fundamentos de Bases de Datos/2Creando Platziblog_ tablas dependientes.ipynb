## creando tabla posts

```mysql
CREATE TABLE `platzi_blog`.`posts` (
  `id` INT NOT NULL AUTO_INCREMENT,
  `titulo` VARCHAR(130) NOT NULL,
  `fecha_publicacion` TIMESTAMP NULL,
  `contenido` TEXT(0) NOT NULL,
  `postscol` CHAR(8) NULL DEFAULT 'activo',
  `usuario_id` INT NOT NULL,
  `categoria_id` INT NOT NULL,
  PRIMARY KEY (`id`));
```

Creando la union entre post y usuarios mediante foreign keys
```mysql
ALTER TABLE `platzi_blog`.`posts` 
ADD INDEX `posts_usuarios_idx` (`usuario_id` ASC) VISIBLE;
;
ALTER TABLE `platzi_blog`.`posts` 
ADD CONSTRAINT `posts_usuarios`
  FOREIGN KEY (`usuario_id`)
  REFERENCES `platzi_blog`.`usuarios` (`id`)
  ON DELETE NO ACTION
  ON UPDATE CASCADE;
```
## creando relacion de tablas posts con categorias 
```mysql
ALTER TABLE `platzi_blog`.`posts` 
ADD INDEX `post_categorias_idx` (`categoria_id` ASC) VISIBLE;
;
ALTER TABLE `platzi_blog`.`posts` 
ADD CONSTRAINT `post_categorias`
  FOREIGN KEY (`categoria_id`)
  REFERENCES `platzi_blog`.`categorias` (`id`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;
```
