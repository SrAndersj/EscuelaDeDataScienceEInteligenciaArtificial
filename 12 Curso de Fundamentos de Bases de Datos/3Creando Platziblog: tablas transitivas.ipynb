uniendo llave foranea comentarios usuarios
```mysql

ALTER TABLE `platzi_blog`.`comentarios` 
ADD INDEX `comentarios_usuarios_idx` (`usuario_id` ASC) VISIBLE;
;
ALTER TABLE `platzi_blog`.`comentarios` 
ADD CONSTRAINT `comentarios_usuarios`
  FOREIGN KEY (`usuario_id`)
  REFERENCES `platzi_blog`.`usuarios` (`id`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;

```

## creacion de tabla transitiva
```mysql
CREATE TABLE `platzi_blog`.`posts_etiquetas` (
  `id` INT NOT NULL AUTO_INCREMENT,
  `post_id` INT NOT NULL,
  `etiqueta_id` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`id`));
```

## llave foranea tabla transitiva a post

```mysql
ALTER TABLE `platzi_blog`.`posts_etiquetas` 
ADD INDEX `postetiquetas_post_idx` (`post_id` ASC) VISIBLE;
;
ALTER TABLE `platzi_blog`.`posts_etiquetas` 
ADD CONSTRAINT `postetiquetas_post`
  FOREIGN KEY (`post_id`)
  REFERENCES `platzi_blog`.`posts` (`id`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;


```
## llave foranea transitoria etiquetas

```mysql

ALTER TABLE `platzi_blog`.`posts_etiquetas` 
ADD CONSTRAINT `postetiquetas_etiquetas`
  FOREIGN KEY (`etiqueta_id`)
  REFERENCES `platzi_blog`.`etiquetas` (`id`)
  ON DELETE NO ACTION
  ON UPDATE NO ACTION;

```
