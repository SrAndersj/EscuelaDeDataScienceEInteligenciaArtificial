# LEFT JOIN
selecciona toda la tabla usuarios 
y unela como izquierda atravez de las dos llaves
## Aqui nos trae al usuario  perezoso que no escribio ningun posts
```mysql

SELECT *
FROM usuarios
	LEFT JOIN posts ON usuarios.id=posts.usuario_id;

```



# LEFT JOIN WHERE
dejame por fuera los que si tienen posts
## aqui nos trae solo al usuario perozoso que no escribio ningun post
```mysql

SELECT *
FROM usuarios
	LEFT JOIN posts ON usuarios.id=posts.usuario_id
    WHERE posts.usuario_id IS NULL;

```

# LEFT RIGTH

## no nos trae al usuario perezoso que no hizo ningun post , pero nos trae a un post que no tiene usuario

```mysql

SELECT *
FROM usuarios
	RIGHT JOIN posts ON usuarios.id=posts.usuario_id;
```

# LEFT RIGTH WHERE 

## nos trae el unico post que no tiene asociado ningun usuario 

```mysql
SELECT *
FROM usuarios
	RIGHT JOIN posts ON usuarios.id=posts.usuario_id
    WHERE posts.usuario_id IS NULL;

```

# INNER JOIN 

## TRAE LA INTERSECCION 
```mysql

SELECT *
FROM usuarios
	INNER JOIN posts ON usuarios.id=posts.usuario_id;

```
