# traeme todos los registros donde el usario sea nulo


```mysql

SELECT *
FROM posts
WHERE usuario_id IS NULL
;




SELECT *
FROM posts
WHERE categoria_id IS NULL
;
```

# traer lo que no sea nulo osea que si tienen usuario id

```mysql

SELECT *
FROM posts
WHERE categoria_id IS NOT NULL
;

```

# filtro sobre filtro sobre filtro
```mysql
SELECT *
FROM posts
WHERE categoria_id IS NOT NULL
	AND estatus = 'activo'	
    
;




SELECT *
FROM posts
WHERE categoria_id IS NOT NULL
	AND estatus = 'activo'	
    AND ID<50
    
;



SELECT *
FROM posts
WHERE categoria_id IS NOT NULL
	AND estatus = 'activo'	
    AND ID<50
    AND categoria_id=2
    
;


SELECT *
FROM posts
WHERE categoria_id IS NOT NULL
	AND estatus = 'activo'	
    AND ID<50
    AND categoria_id=2
    AND YEAR(fecha_publicacion)='2025'
    
;

```
