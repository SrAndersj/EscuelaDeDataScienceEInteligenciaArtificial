# inversa 

es la matriz que al operar matriz A por inversa da identidad


$A A^{-1} = Id$


```python
A = np.array([[1,0,1],[0,1,1],[-1,1,1]])

A
```

```python
inversa_A=np.linalg.inv(A)
inversa_A
```

```python
A.dot(inversa_A)
```

no siempre existe esta identidad y si no existe la llamamos matriz singular

```python
singular = np.array([[1,1],[1,1]])
singular
```

```python
np.linalg.inv(singular)
```
