````python
import plotly.graph_objects as go
import numpy as np

# Creamos una lista con todas las coordenadas x e y posibles
x_coords = np.arange(-10*v1[0], 10*v1[0], v1[0]) + np.arange(-10*v2[0], 10*v2[0], v2[0])[:, np.newaxis]
y_coords = np.arange(-10*v1[1], 10*v1[1], v1[1]) + np.arange(-10*v2[1], 10*v2[1], v2[1])[:, np.newaxis]

# Creamos una figura de Plotly
fig = go.Figure()

# Añadimos los puntos a la figura
fig.add_trace(
    go.Scatter(
        x=x_coords.flatten(),
        y=y_coords.flatten(),
        mode='markers',
        marker=dict(
            color='orange',
            size=5,
        ),
    )
)

# Agregamos las líneas verticales y horizontales
fig.add_shape(
    dict(
        type="line",
        x0=0,
        y0=-25,
        x1=0,
        y1=25,
        line=dict(color="grey")
    )
)
fig.add_shape(
    dict(
        type="line",
        x0=-25,
        y0=0,
        x1=25,
        y1=0,
        line=dict(color="grey")
    )
)

# Establecemos los límites del eje x e y
fig.update_xaxes(range=[-25, 25])
fig.update_yaxes(range=[-25, 25])

# Mostramos la figura
fig.show()


````

````python
from mpl_toolkits.mplot3d import Axes3D

v1 = np.array([1, 0, 0])
v2 = np.array([2, -3, 0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   v1[2]*a + v2[2]*b,
                   marker = '.',
                   color = "orange")
        
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
````
