```python
A = np.diag([2,3,4,5])
print(A)
```

```python
v1 = np.array([[1,1,1,1]])

print(v1)

print(' ')

print(A.dot(v1.T))
```

```python
A_inv = np.diag([1/2,1/3,1/4,1/5])
print(A_inv)
```

```python
identidad = A.dot(A_inv)

print(identidad)
```

```python
A_inv_calc = np.linalg.inv(A)
print(A_inv_calc)
```
