Una matriz es ortogonal cuando todas sus filas son mutuamente ortonormales y todas sus columnas son mutuamente ortonormales

```python
import numpy as np
```

```python
matriz = np.array([[1,0,0],
                   [0,1,0],
                   [0,0,1]
                  ])
print(matriz)
```

para ver que es ortogonal esta matriz necesitamos comprobar que sus columnas son mutuamente ortonormales



```python
# primera columna
matriz[:,0]
```

```python
# segunda columna
matriz[:,1]
```



$$
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$


$$
\begin{bmatrix}
v_1 & v_2 & v_3 \\
 &  &  \\
 &  & 
\end{bmatrix}
$$




para comprobar la ortogonalidad del vector 1(columna 1) respecto a v2

 $v_1 \perp v_2$

```python

# la primer columna con el proudcto interno con la segunda columna
matriz[:,0].dot(matriz[:,1])
```

 $v_1 \perp v_3$

```python
matriz[:,0].dot(matriz[:,2])
```

$v_2 \perp v_3$

```python
matriz[:,1].dot(matriz[:,2])
```

para estar seguros que son ortonormales validamos su norma

```python
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
```

```python
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
```

ver la norma con respecto filas

```python
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
```

```python
A = np.array([[np.cos(100), -np.sin(100)],
             [np.sin(100), np.cos(100)]])
print(A)
```

queremos comprobar que esta matriz es ortogonal

```python
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
```

empezamos a ver los problemas para representar computacionalmenbte , este numero deberia ser 1 pero es 0.999....

veamos que el angulo que forman entre ellos es cero

producto interno de una fila con la otra

y el producto interno de una columna con la otra

```python
print(A[0,:].dot(A[1,:]))
print(A[:,0].dot(A[:,1]))
```

ahora queremos ver que 

```python
A_t = A.T
print(A_t.dot(A))
print(A.dot(A_t))
```
