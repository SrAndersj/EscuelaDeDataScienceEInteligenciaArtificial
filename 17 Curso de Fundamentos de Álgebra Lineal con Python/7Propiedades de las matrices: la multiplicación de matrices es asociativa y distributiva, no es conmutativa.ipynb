```python
import numpy as np
```

```python
A = np.array([[2,3],[5,7],[11,13]])
B = np.array([[1,3],[2,1]])
C = np.array([[3,1],[4,2]])
```

# Asociativa

```python

ABC = A.dot(B.dot(C))
print(ABC)
```

```python
AB_C = A.dot(B).dot(C)
print(AB_C)
```

#Distrbutiva

```python

D = A.dot(B+C)
E = (A.dot(B))+(A.dot(C))
```

```python
print(D)
print('')
print(E)
```

```python
print(D==E)
```
