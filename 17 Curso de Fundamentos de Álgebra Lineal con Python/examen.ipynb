## ¿Qué ángulo forman los vectores v1 = [0,1] y v2 = [1,0]?



```python
import numpy as np

# Definir los vectores
v1 = np.array([0,1])
v2 = np.array([1,0])

# Calcular el producto punto
dot_product = v1.dot(v2)

# Calcular la magnitud de los vectores al cuadrado
v1_mag_squared = np.linalg.norm(v1)**2
v2_mag_squared = np.linalg.norm(v2)**2

# Calcular el ángulo en radianes utilizando la fórmula
angle = np.arccos(dot_product / np.sqrt(v1_mag_squared * v2_mag_squared))

# Convertir el ángulo de radianes a grados
angle_degrees = np.degrees(angle)

# Imprimir el ángulo en grados
print("El ángulo entre los vectores v1 y v2 es de:", angle_degrees, "grados")
```

# ¿Las matrices singulares tienen inversa?

## No, las matrices singulares no tienen inversa. Una matriz singular es aquella que no tiene inversa, es decir, no se puede encontrar una matriz que, multiplicada por la matriz singular, produzca la matriz identidad.

Geométricamente, una matriz singular representa una transformación lineal que comprime o aplana el espacio en una o más direcciones, y por lo tanto, no puede ser invertida en esas direcciones.

Por otro lado, una matriz no singular es aquella que sí tiene inversa, es decir, se puede encontrar una matriz que, multiplicada por la matriz no singular, produzca la matriz identidad.

En resumen, las matrices singulares no tienen inversa y no pueden ser invertidas, mientras que las matrices no singulares sí tienen inversa y pueden ser invertidas.

# Un sistema de ecuaciones lineales en R2 tiene infinitas soluciones cuando: 

 Un sistema de ecuaciones lineales en R2 tiene infinitas soluciones cuando las dos ecuaciones lineales representan la misma recta. En Python, podemos verificar esto usando la librería NumPy y el siguiente código:

````python
import numpy as np

# Definimos las matrices A y B que representan las ecuaciones lineales
A = np.array([[1, -2],
              [2, -4]])

B = np.array([3, 6])

# Resolvemos el sistema de ecuaciones lineales
x = np.linalg.solve(A, B)

# Imprimimos la solución
print("Solución del sistema de ecuaciones lineales:")
print(x)

````

# ¿Cómo puedo comprobar si una matriz A es simétrica?


Una matriz A es simétrica si y solo si es igual a su propia traspuesta, es decir, si se cumple que A = A^T. Por lo tanto, para comprobar si una matriz A es simétrica, podemos simplemente verificar si se cumple esta igualdad.

En Python, podemos utilizar la función numpy.allclose() para comparar si dos matrices son iguales con una tolerancia dada. Por lo tanto, podemos comprobar si una matriz A es simétrica con el siguiente código:

````python
A ==A.T
````

# Dadas A y B, tal que A.dot(B) está definido, entonces al transponer(A.dot(B)) siempre es igual a:

````python
import numpy as np

# Definimos las matrices A y B
A = np.array([[1, 2], 
              [3, 4], 
              [5, 6]])
B = np.array([[7, 8, 9], 
              [10, 11, 12]])

# Multiplicamos las matrices A y B
C = A.dot(B)

# Imprimimos la matriz C y su transpuesta
print("Matriz C:\n", C)
print("Transpuesta de C:\n", C.T)

# Calculamos B.T.dot(A.T)
D = B.T.dot(A.T)

# Imprimimos la matriz D
print("Matriz D:\n", D)
````
