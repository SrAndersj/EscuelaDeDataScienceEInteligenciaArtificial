```python
def graficarVectores(vecs, cols, alpha = 1):
    
    plt.figure()
    plt.axvline(x = 0, color = "grey", zorder = 0)
    plt.axhline(y = 0, color = "grey", zorder = 0)
    
    for i in range(len(vecs)):
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', scale_units='xy', scale=1,
                   color=cols[i],
                  alpha = alpha)
```

```python
def graficarVectoresText(vecs, cols, alpha=1):
    plt.figure()
    plt.axvline(x=0, color="grey", zorder=0)
    plt.axhline(y=0, color="grey", zorder=0)
    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
        plt.text(x[2], x[3], f"({x[2]}, {x[3]})", fontsize=12, color=cols[i], verticalalignment='bottom', horizontalalignment='left')
    plt.xlim(-1,2)
    plt.ylim(-1,2)
```

# la misma grafica uno pero en 3D



```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def graficarVectores3D(vecs, cols, alpha=1):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlim([0, np.max(vecs)])
    ax.set_ylim([0, np.max(vecs)])
    ax.set_zlim([0, np.max(vecs)])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    for i in range(len(vecs)):
        x = np.concatenate([[0], vecs[i]])
        ax.quiver(0, 0, 0, x[1], x[2], x[3], color=cols[i], alpha=alpha)
        
    plt.show()
```
