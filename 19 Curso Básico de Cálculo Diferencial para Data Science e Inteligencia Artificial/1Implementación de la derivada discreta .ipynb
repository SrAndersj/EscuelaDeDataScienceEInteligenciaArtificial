$$ f(x) = x^2 $$

$$ f'(x) = \lim_{h\to0} \frac{f(x+h) - f(x)}{h} $$ 

.

.

$$ f'(x) = \lim_{h\to0} \frac{(x+h)^2 - x^2}{h} $$ 

.

.

\begin{align*}
f'(x) &= \lim_{h\to0} \frac{x^2 + 2xh + h^2 - x^2}{h} \\
      &= \lim_{h\to0} \frac{2xh + h^2}{h} \\
      &= \lim_{h\to0} (2x + h) \\
      &= 2x
\end{align*}


.

.

$$ f'(x) = 2x $$

```python

# Definición de la función
def f(x):
  return x**2+3+x**4

# Creación del array de valores de x
x = np.linspace(-10, 10, num=1000)

# Evaluación de la función para cada valor de x
y = f(x)

# Definición de la función derivada
def df(f):
  h = 0.00001
  return (f(x+h) - f(x)) / h

# Gráfica de la función
plt.plot(x, y, 'b', label='función')

# Gráfica de la derivada de la función
plt.plot(x, df(f), 'r', label='derivada')

# Agregar leyendas a la gráfica y ubicarlas en la esquina superior derecha
plt.legend(loc='upper right')

# Mostrar la gráfica
plt.show()

```
