# MSE 

## compara valor real con el predicho

* a valores grandes , peor se agusta el modelo
* problema de dimensionalidad


La ecuación del modelo MSE es: 

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y_i})^2$$

Donde:
- $n$: número de observaciones
- $y_i$: valor real de la i-ésima observación
- $\hat{y_i}$: valor predicho para la i-ésima observación


# $R^2$

## intenta mostrar que tambien se ajusta el modelo en base a la varianza del modelo

La ecuación del coeficiente de determinación R cuadrado en términos de sumas es:

$$R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y_i})^2}{\sum_{i=1}^n (y_i - \bar{y})^2}$$

Donde:
- $n$: número de observaciones
- $y_i$: valor real de la i-ésima observación
- $\hat{y_i}$: valor predicho para la i-ésima observación
- $\bar{y}$: media de los valores observados
