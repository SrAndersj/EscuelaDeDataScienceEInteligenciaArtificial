```python
import pandas as pd
import numpy as np
```

# vamos a crear un peque√±o diccionario

```python
dict= {'col1':[1,2,3,np.nan],
 'col2':[4,np.nan,6,7],
 'col3':['a','b','c',None] 
 }
 
```

```python
df=pd.DataFrame(dict)
df
```

# manejo de datos nulos 
## que si es nulo , donde encontremos un nan el va a decir Ture

```python
df.isnull()
```

# y esto lo podemos llevar a un metodo numerico que se usa mucho en tensores

```python
df.isnull()*1
```

# una vez identificados vamos a usar fillna para llenar los valores nulos con la palabra Missing esto con el fin de despues poder usar filtros


## 

```python
df.fillna('Missing')
```

# Filna puede llenar los valores con la media

```python
df.fillna(df.mean())
```
