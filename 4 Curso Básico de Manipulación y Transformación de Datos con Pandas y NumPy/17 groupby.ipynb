```python
import pandas as pd
```

```python
df_books=pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',')
df_books.head(2)
```

# uso de agrupamientos o funciones de agregacion

```python
df_books.groupby('Author')
```

# vamos a saber cuantos por author

```python
df_books.groupby('Author').count()
```

# puedo saber el maximo , el minimo

```python
df_books.groupby('Author').max().head(2)
```

```python
df_books.groupby('Author').min().head(2)
```

# author esta un poco mas abajo de las comunas entonces eso es porque es mi indice del dataframe

```python
df_books.groupby('Author').count().loc['Walter Isaacson']
```

# ahora que pasa si yo quiero los valores planos

osea lo que salio pero en un dataframe


```python
df_books.groupby('Author').count().reset_index()
```

```python
df_books.groupby('Author').sum().reset_index()
```

# pdemos hacer que el agrupamiento sea por determinadas unciones de agregacion 


```python
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.groupby('Author').agg(['count','sum'])
```

# ahora quiero que de los reviews haga maximo y minimo pero del usser rating haga suma

```python
df_books.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Year']).count()
```

```python
df_books.groupby(['Author','Year']).count().reset_index()
```
