# merge

el merge

```python
izq =pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})

der =pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
izq
```

```python
der
```

# combinacion entre izquierda y derecha
## y pandas lo esta haciendo por las columnas que tienen en comun

```python
izq.merge(der)
```

# vamos a hacer el merge sobre una columna especifica, key


```python
izq.merge(der,on='key')
```

# voy a cambiar el nombre key  por key_2 y va a fallar cuando solicito que hagar merge poer medio de key

```python
izq =pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})

der =pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
izq.merge(der,on='key')
```

# entonces solucionamos idicandole que por un lado tal y el otro tal

```python
izq.merge(der,left_on='key',right_on='key_2')
```

# cuando ponemos un valor en nan no concatena 


```python
izq =pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7']})

der =pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
der
```

```python
izq.merge(der,left_on='key',right_on='key_2')
```

# el comando how nos va a salvar la vida y permitir que la del NaN si se una

inner es el valor por default


```python
izq.merge(der,left_on='key',right_on='key_2',how='left')
```

```python
izq.merge(der,left_on='key',right_on='key_2',how='inner')
```

```python
izq.merge(der,left_on='key',right_on='key_2',how='right')
```
