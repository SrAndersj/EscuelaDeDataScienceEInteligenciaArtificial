```python
import pandas as pd
import numpy as np
```

# JOIN index match

```python
izq=pd.DataFrame({'A':['A4','A5','A6'],
 'B':['B4','B5','B6']},index=['k0','k1','k2'])

der=pd.DataFrame({'C':['C4','C5','C6'],
 'D':['D4','D5','D6']
 },index=['k0','k2','k3'])
```

```python
izq.join(der)
```

# si ponemos inner , hacen match los que estan en los dos

outer es traigame todos los datos


```python
izq.join(der,how='inner')
```

```python
izq.join(der,how='outer')
```

```python
izq.join(der,how='left')
```
