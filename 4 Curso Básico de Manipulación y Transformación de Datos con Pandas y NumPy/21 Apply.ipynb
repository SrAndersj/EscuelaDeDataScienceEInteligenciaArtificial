```python
import pandas as pd
```

# Nos ayuda a aplicar funciones a nuestro dataframe


```python
df_books = pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv',sep=',',header=0)

df_books.head()
```

# vamos a crear una funcion de python

una funcion que se llame  two times con un value 

```python
def two_times(value):
  return value *2
```

# apply va a tomar cada uno de los valores de cada fila de la columna y le va a aplicar la funcion

```python
df_books['User Rating'].apply(two_times)
```

# creamos una nueva columna
con esos datos

```python
df_books['Rating_2']=df_books['User Rating'].apply(two_times)
```

```python
df_books.head()
```

# podemos aplicar cualquier tipo de funcion

como funciones de lambda


```python
df_books['User Rating'].apply(lambda x: x*3)
```

# que pasa si le mando todas las columnas  y no solo una , pues puedo usar hasta sentencias if


si el genero es fiction multiplique por dos de no ser asi dejelo igual

# da error porque no decimos que axis son los que usaremos de indices

```python
df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] =='Fiction' else x['User Rating']  )
```

# especificamos el axis y ya


```python
df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] =='Fiction' else x['User Rating'],axis=1  )
```

```python
df_books['Rating_2']=df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] =='Fiction' else x['User Rating'],axis=1  )
```
