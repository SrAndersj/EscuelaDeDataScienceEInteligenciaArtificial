```python
import numpy as np
```

```python
arr= np.array([1,2,3,4])
```

```python
arr.dtype
```

# definir dtype

```python
arr= np.array([1,2,3,4], dtype='float64')
arr.dtype
```

# cambiarlo si ya esta definido

```python
arr= np.array([1,2,3,4])
arr.dtype
```

```python
arr=arr.astype(np.float64)
arr.dtype
```

# ahora vamos a convertirlo en bool 

```python
arr= np.array([0,1,2,3,4])
arr.dtype
```

```python
arr=arr.astype(np.bool_)
arr.dtype
```

```python
arr
```

#ahora lo convertimos a string

```python
arr= np.array([0,1,2,3,4])
arr=arr.astype(np.string_)
arr
```

vamos a pasar de string a entero int8

```python
arr= np.array(['0','1','2','3','4'])
arr=arr.astype(np.int8)
arr
```
