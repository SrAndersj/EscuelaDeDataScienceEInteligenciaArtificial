# Matriz

```python
matriz =np.array([[1,2,3],[4,5,6]])
print(matriz)

print('numero de dimensiones ') 
matriz.ndim
```

```python
tensor =np.array([[[1,2,3],[4,5,6]],[[6,7,8],[0,7,5]]])
print(tensor)

print('numero de dimensiones ') 
tensor.ndim
```

# agregaar o eliminar dimensiones 
ndmin =10  significa que el numero de dimensiones desde que lo defino tiene que ser 10 

```python
vector =np.array([1,2,3],ndmin=10)
print(vector)
vector.ndim
print('numero de dimensiones ') 
vector.ndim

#agr
```

# ahora si tengo un array y quiero expandierle una unica dimension en sus ejes

np.expand_dims

axis =0  significa que voy a expandir una dimension a nivel de filas

```python
expand = np.expand_dims(np.array([1,2,3]),axis=0)
print(expand)
print('numero de dimensiones ') 
expand.ndim
```

# cuando quiero eliminar dimensiones que no estoy usando 

np.squeeze  comprime al numero correco de dimensiones que se esta usando , osea la que se cree correcta 

```python
print(vector,vector.ndim)

vector_2=np.squeeze(vector)
print(vector_2,vector_2.ndim)
```
