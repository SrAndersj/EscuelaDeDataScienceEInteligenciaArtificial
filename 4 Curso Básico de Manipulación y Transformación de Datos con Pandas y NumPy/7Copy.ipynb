```python
import numpy as np
```

a veces se puede hacer modificaciones que fallan y dañan el original entonces se usa copy para evitar perder 

```python
arr =np.arange(0,11)
arr
```

# vamos a hacer un slicing con las posiciones de la 0 a la 6

```python
trozo_de_arr=arr[0:6]
trozo_de_arr
```

# vamos a convertir todos los valores del trozo a cero

```python
trozo_de_arr[:]=0
trozo_de_arr
```

# ahora cuando veo** arr** noto que esta dañado , error fatal

```python
arr
```

# este problema lo puedo evitar  con copy

```python
arr_copy = arr.copy()
arr_copy
```

```python
arr_copy[:]=100
arr_copy
```

```python
arr
```

# siempre que queremos usar un trozo de un array debemos usar copy para no dañar el original
