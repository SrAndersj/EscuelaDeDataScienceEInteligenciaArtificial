```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# seaborn tiene datos pregargados 

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='tip')
plt.show()
```

# Relacion entre el total de la cuenta y la propina

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

# hue me hace un desgloce o agrupamiento por cualquier variable

entonces la relacion entre x y la va a separar por colores

```python
sns.displot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

# kind , kde es que es la misma distribucion pero por densidad

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=False)
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True)
plt.show()
```

# paleta de colores

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark')
plt.show()
```

# Transparencia


```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark',alpha=.25)
plt.show()
```
