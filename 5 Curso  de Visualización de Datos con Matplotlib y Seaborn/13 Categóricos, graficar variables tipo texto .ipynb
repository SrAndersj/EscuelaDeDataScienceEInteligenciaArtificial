```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
tips.head(2)
```

# vamos a contar la variable categorica day  en la cual puedo tener 

jueves Thur

viernes Fri

```python
sns.countplot(data=tips,x='day',hue='sex')
```

```python
sns.countplot(data=tips,y='day',hue='sex')
```

# stripplot se usa para 

```python
sns.stripplot(data=tips,x='day',hue='sex')
plt.show()
```

# le especifico que quiero ver de esas variables categoricas

vamos a ver por dia como es la dispersion del total de la cuenta para hombres y mujeres

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# swarmplot en esta ocasion es similar solo que en la concentracion de datos voy a tener un eje horizontal

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# vamos a usar el boxplot

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# boxplot tambien sirve para graficar variables numericas y no solo categoricas

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,y='total_bill')
plt.show()
```

# vamos  a combinar

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# cambiemosle el color solo al swarmplot

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0") # color 0 es negro
plt.show()
```

# marker cambiado

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0",marker='<') # color 0 es negro
plt.show()
```

# violinplot es muy parecido al boxplot si no que  me va a decir como se estan concentrando los datos 

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',palette='pastel')
plt.show()
```

# ahora con split los combinamos en un grafico y no en dos

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
```

# catplot grafica variables categoricas

sin siquiera mencionarle el grafico que quiero trabajar el va a definir que quiere usar un strip

lo interesante es que es paecido a displot si le digo con kind que quiero usar otro grafico funciona

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# lo interesante es que es paecido a displot si le digo con kind que quiero usar otro grafico funciona

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='swarm')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box')
plt.show()
```

# vamos a agregar otra columna

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time')
plt.show()
```
