```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# ver correlacion entre distintas variables numericas

```python
tips=sns.load_dataset("tips")
tips.head(5)
```

# hacemos un grafico de distribuciones , con un scatterplot

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# observamos que parece haber una correlacion no muy fuerte porque parece que entre mas crece la cuenta mas es la propina

## ahora vamos a segmentarlo por el dia

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day')
plt.show()
```

# vamos a decirle que el estilo dependa de la variable time

time dice si es Dinner o launch

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time')
plt.show()
```

# vamos a cambiar el tamao del marcador por el tamao del numero de los comensales size del dataframe

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.show()
```

# reacomodar la leyenda

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()

```

# propiedades utiles

digamos que los markers , si el marker es lunch use un diamante


para Dinner vamos a usar un cuadrado(square)

```python


markers={"Lunch":"D","Dinner":"s"}

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,8))

markers={"Lunch":"D","Dinner":"s"}

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

# Tenemos un lineplot

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
```

# replot

es como el catplot y el displot


cambiamos usando el kind

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers,kind='line')
```

# vamos a usar col

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers,col='time')
```
