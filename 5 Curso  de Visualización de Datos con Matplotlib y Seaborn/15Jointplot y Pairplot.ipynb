```python
import seaborn as sns
import matplotlib.pyplot as plt
```

nos ayuda mucho porque con un solo comando combino varios tipos de grafico


```python
tips=sns.load_dataset("tips")
tips.head(5)
```

# unio distintas graficas en un solo valor

```python
sns.jointplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# vamos a segmentarlos por sex

lo chevere es que sns cambia de manera automatica a la forma que cree que es mejor para graficar

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

# kind

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='kde')
plt.show()
```

# Marginal para agregar un pequeo eje que dice  me dice la escala de estos histogramas

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True)
plt.show()
```

# marginals kws para afectar los histogramas de los ejes y no a la relacion

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```
