```python
import seaborn as sns
import matplotlib.pyplot as plt

```

# Heatmap es para graficar datos de estructura matricial

```python
tips=sns.load_dataset("tips")
tips.head(5)
```

# corr()
# vamos a llevar esto a una estructura matricial especial

## corr es una funcionalidad de pandas para obtener la correlacion dentro de las variables que va desde -1 a 1 

### donde  -1  es una correlacion indirecta , 0 es que no hay y 1 correlacion directa

# correlacion es una forma de decir que un grupo de datos aumenta o disminuye en funcion de los otros datos 

```python
tips.corr()
```

# si tuvieramos 100 variables seria muy aburrido trabajar con el metodo corr



# entonces sns crea heatmap

puedo ver con gamas de colores como se correlacionan la variables

```python
sns.heatmap(tips.corr())
plt.show()
```

# con annot ahora tengo tambien el valor de la correlacion

```python
sns.heatmap(tips.corr(),annot=True)
plt.show()
```

# vamos a cambiar el mapa de colores


```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5)
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black')
plt.show()
```

# vamos a decirle que el valor minimo va ser igual a 0 1 y entonces la escla de colores va de cero a 1



```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
 vmin=0,vmax=1
 
 )
plt.show()
```

# Eliminar la barra

```python
sns.heatmap(tips.corr(),annot=True,cbar=False)
plt.show()
```
