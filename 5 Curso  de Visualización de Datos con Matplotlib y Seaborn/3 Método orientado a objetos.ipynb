```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

# vamos a definir una figura o un lienzo

## las graficas que vamos a pintar se llaman axes

a la figura le vamos a agregar axes con add_axes

## luego le decimos que queremos hacer un plot

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.5,0.9])

axes.plot(x,y,'b')

fig.show()
```

# ahora no quiero solo un lienzo dentro de esta figura

## quiero un segundo grafico


                       los dos primeros hacer referencia a la posicion en el lienzo los siguientes dos hace referencia al tamaño de ese axes
axes=fig.add_axes([0.1  ,0.1,   0.5,     0.9])

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.9])
axes2=fig.add_axes([0.1,0.1,0.4,0.3])



axes.plot(x,y,'b')

axes2.plot(y,x,'r')

fig.show()
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.9])
axes2=fig.add_axes([0.15,0.6,0.4,0.3])



axes.plot(x,y,'b')

axes2.plot(y,x,'r')
axes2.set_facecolor('gray')

fig.show()
```
