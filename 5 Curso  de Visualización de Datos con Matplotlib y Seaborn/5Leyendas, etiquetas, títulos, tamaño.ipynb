```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig,axes=plt.subplots(1,2)
```

```python
fig,axes=plt.subplots(nrows=1 ,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Relacion x - y ')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x ')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')


```

# la figura esta un poco pequeña vamos a cambiar el figsize


```python
fig,axes=plt.subplots(nrows=1 ,ncols=2,figsize=(5,5))
axes[0].plot(x,y)
axes[0].set_title('Relacion x - y ')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x ')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
fig,axes=plt.subplots(nrows=1 ,ncols=2,figsize=(10,5))
axes[0].plot(x,y,label="sin(x)")
axes[0].set_title('Relacion x - y ')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()# esto es para que salga el label que puse arriba en plot

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x ')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

# ahora un truco , dejando el simbolo** $**  da notacion matematica

```python
fig,axes=plt.subplots(nrows=1 ,ncols=2,figsize=(10,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion x - y ')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()# esto es para que salga el label que puse arriba en plot

axes[1].plot(y,x)
axes[1].set_title('Relacion y - x ')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

# como hacerlo sin usar el metodo orientado a objetos

```python
plt.plot(x,y,label='sin(x)')
plt.title('este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left',bbox_to_anchor=(0.2, 0.3))
```
