# modificaciones de tipo matlab

```python
plt.style.use('dark_background')
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

# hay otra forma con pyplot

```python
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green')
ax.plot(x,x+2,color='blue')
ax.plot(x,x+3,color='#66FF66')  # notacion rgb
ax.plot(x,x+4,color='#CCFFCC')
```

# nivel de transparencia con alpha y grosor linewidth


```python
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green',alpha=0.5,linewidth=5)
ax.plot(x,x+2,color='blue',alpha=0.2,linewidth=8)
ax.plot(x,x+3,color='#66FF66',linewidth=3)  # notacion rgb
ax.plot(x,x+4,color='#CCFFCC',linewidth=2)
```

# estilo


```python
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green',alpha=0.5,linewidth=5,linestyle='-')
ax.plot(x,x+2,color='blue',alpha=0.2,linewidth=8,linestyle='--')
ax.plot(x,x+3,color='#66FF66',linewidth=3,linestyle='dashed')  # notacion rgb
ax.plot(x,x+4,color='#CCFFCC',linewidth=2,linestyle=':')
```

# Marcadores

```python
fig,ax=plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green',linewidth=5,linestyle='-',marker='o',markersize=10,markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color='blue',linewidth=8,linestyle='--',marker='x')
ax.plot(x,x+3,color='#66FF66',linewidth=3,linestyle='dashed',marker='v')  # notacion rgb
ax.plot(x,x+4,color='#CCFFCC',linewidth=2,linestyle=':',marker='v')
```
