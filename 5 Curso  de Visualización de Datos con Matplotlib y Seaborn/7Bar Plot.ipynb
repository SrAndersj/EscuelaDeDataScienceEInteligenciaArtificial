```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country=['INDIA','JAPAN','MEXICO','COLOMBIA','GEMANY']
population=[1000,800,900,1000,300]
```

```python
plt.bar(country,population)
```

```python
plt.bar(country,population,width=0.5)
plt.show()
```

```python
plt.bar(country,population,width=0.5,color=['red','blue'])
plt.show()
```

# para cambiar el nombre a español

```python
plt.bar(country,population,width=0.5,color=['red','blue'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)# un arreglo de 5 indices
plt.show()
```
