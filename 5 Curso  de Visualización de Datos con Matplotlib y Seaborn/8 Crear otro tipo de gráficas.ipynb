```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data =np.random.randint(1,50,100)
```

## histograma

```python
plt.hist(data)
plt.show()
```

## Boxplot

```python
plt.boxplot(data)
plt.show()
```

# que pasa se ingreso data anomala


el rango 200 que esta tan lejos 

lo deja separado

```python
data=np.append(data,200) # agregamos un dato lejisimos

plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

# si quiero que no me muestre estos outlayers uso showfliers

```python
data=np.append(data,200) # agregamos un dato lejisimos

plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

# veamoslo en True osea con todos los datos , osea los outlayers en este caso el 200


```python
data=np.append(data,200) # agregamos un dato lejisimos

plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

# Scatter  dispersion

es la grafica que me muestra la relacion entre dos distinas variables


se utiliza para ver si hay correlacion o no entre las variables

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

area= (30* np.random.rand(N) **2)

colors=  np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

# que pasa si le agrego una variable extra como area , de manera que el tamaño que tengo en cada una  de las burbujas hace referencia a area


burbujas mas grandes 

```python
plt.scatter(x,y,s=area)
plt.show()
```

# que pasa si meto otra variable mas , si meto color



```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

# puedo cambiar el marker


```python
plt.scatter(x,y,s=area,c=colors,marker='x')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='<')
plt.show()
```

Alpah

```python
plt.scatter(x,y,s=area,c=colors,marker='<',alpha=0.5)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='p',alpha=0.5)
plt.show()
```
