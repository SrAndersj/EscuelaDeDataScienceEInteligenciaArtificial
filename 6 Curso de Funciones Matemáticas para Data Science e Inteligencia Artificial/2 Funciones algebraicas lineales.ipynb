### Funci√≥n lineal

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$


```python
N=100  # resolucion o puntos que vamos a ver en la grafica

m= 10

b=5

def f(x):
  return m*x + b

x=np.linspace(-10,10,num=N) # valores de -10 a 10 y que sean 100 valores con num 

y=f(x)

fig,ax =plt.subplots()
ax.plot(x,y)


```
