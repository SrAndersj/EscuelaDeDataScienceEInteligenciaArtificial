# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y= np.zeros(len(x)) # variable Y temporal que va a ser puros 0 que tengamos de la longitud len de la variable x
  for idx,x in enumerate(x):             # por cada indice de x idx y x in enumerate que nos permite obtener el indice y el elemento de un arreglo en python
    if x >= 0:
      Y[idx]=1
  return Y


N=1000

x=np.linspace(-10,10,num=N)

y=H(x)

plt.plot(x,y)
```
