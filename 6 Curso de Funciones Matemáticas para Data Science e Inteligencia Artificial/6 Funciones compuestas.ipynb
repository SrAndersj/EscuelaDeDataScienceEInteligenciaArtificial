```python
import matplotlib.pyplot as plt
import numpy as np
```

conosidas las funciones f y g la composicion de f y g esta dada por:

$$ f \circ g =(f \circ g)(x)=f(g(x)) $$





```python
N=1000

x=np.linspace(-10,10,num=N)

def g(x):
  return x**2

def f(x):
  return np.sin(x)


plt.plot(x,g(x))
```

```python

plt.plot(x,f(x))
```

```python
f_o_g = f(g(x))

plt.plot(x,f_o_g)
```
