```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

#### Leemos la base de datos a trabajar en nuestra ruta de origen

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
```

```python
iris.head()
```

flor.jpeg

```python
plt.hist(iris['sepal-length'],color = 'orange',bins =int(180/5))
```

# queremos estar seguros de que si es una distribucion normal 


kernel densite estimate kde que es una linea que una nuestra distribucion de puntos

```python
sns.distplot(iris['sepal-length'], hist=True, kde=True, 
             bins=int(180/5), color = 'orange')
```

#### Clculo de la varianza

$ \sigma ^{2} = \sum_{i=1}^{N} \frac{()x_i-\mu)^{2}}{N }$

```python
iris["sepal-length"].var()
```

 #### desviacion estandar

$ \sigma =\sqrt{\sigma} ^{2} = \sum_{i=1}^{N} \frac{()x_i-\mu)^{2}}{N }$

```python
iris["sepal-length"].std()
```

vamos a sacar el promedio para analizar y comparar


si por ejemplo el mean diera 100 implicaria que las desviaciones serian cerca del uno , entonces indicaria que hay poca desviacion 


en el caso actual con mean 5.84 y desviacioens de 0.68 y la de 0.82 concluimos que la desviocion es grande

```python
iris["sepal-length"].mean()
```
