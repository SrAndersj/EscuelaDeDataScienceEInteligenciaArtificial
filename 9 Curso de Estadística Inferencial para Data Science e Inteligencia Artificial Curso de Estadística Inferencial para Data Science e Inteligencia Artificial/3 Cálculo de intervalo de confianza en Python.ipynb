```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import scipy.stats as st
```

#### Esta práctica la haremos sobre una distribución creada por nosotros

```python
np.random.seed(20) 
var1 = np.random.randint(0, 10, 10) # Días de vacaciones
var2 = var1+np.random.normal(0, 1, 10) # Dinero gastado
```

```python
var1
```

```python
var2
```

#### Visualización de resultados con un intervalo de confianza del 80%


regplot es un plot de regresion

ci es coeficiente interval



```python
grafica = sns.regplot(var1,var2,ci=80)
```

# vamos a crear una tupla con los datos

```python
tuple =list(zip(var1,var2))
tuple
```

```python
tabla = pd.DataFrame(tuple,columns=['Días_vacaciones','Dinero_gastado'])
tabla
```

# vamos a calcular los intervalos de confianza a u n 95 porciento


```python
st.t.interval(alpha=0.95,df=len(tabla)-1,loc = np.mean(tabla),scale=st.sem(tabla))
```
