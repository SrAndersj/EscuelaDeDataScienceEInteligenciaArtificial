```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

import scipy.stats as st


from scipy.stats import f_oneway

from scipy.stats import ttest_ind
```

#### De nuevo, usaremos la base de datos Iris

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
iris.head()
```

## Tipos de pruebas:

### a) Distribuci√≥n t de Student



# $ SE =\frac{\sigma}{\sqrt{N}} $

SE = standart error

.std = standart desviation =$\sigma$

```python
se_length =iris.std()['sepal-length']/np.sqrt(10)  

se_width = iris.std()['sepal-width']/np.sqrt(10)
```

# SED = standart error desviation

# $SED =\sqrt{ \frac{\sigma_{1}^{2}}{N_1}+ \frac{\sigma_{1}^{2}}{N_2}} $

```python
sed = np.sqrt((se_length**2)+se_width**2)
```

# $t = \frac{\bar{x_1}-\bar{x_2}}{SED}    $  


# $ t = \frac{x_1-x_2}{\frac{(s_{1})^2}{n_1}+\frac{(s_{2})^2}{n_2}}    $




```python
t_stat =(iris.mean()['sepal-length']-iris.mean()['sepal-width'])/sed


t_stat
```

# vamos a caluclar el valor de p

```python
p =ttest_ind(iris['sepal-length'],iris['petal-width'])

p
```
