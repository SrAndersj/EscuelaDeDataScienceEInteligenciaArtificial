```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)


data = pd.DataFrame({'x': x, 'y': y})


data.plot.scatter(x='x', y='y')
plt.show()
```
