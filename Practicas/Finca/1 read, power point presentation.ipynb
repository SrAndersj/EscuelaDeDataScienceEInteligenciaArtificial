```python

!pip install python-pptx
from pptx import Presentation


X = Presentation()

Layout = X.slide_layouts[0] 

first_slide = X.slides.add_slide(Layout) # Adding first slide


first_slide.shapes.title.text = "Bitacora"

first_slide.placeholders[1].text = str(df['new_date'][0])

X.save("First_presentation.pptx")
```

# siguiente pagina , y cuadro de texto


```python
#En primer lugar, deberá importar métodos adicionales para agregar contenido.
from pptx.util import Inches

# Primero creemos y agreguemos la segunda diapositiva.

Second_Layout = X.slide_layouts[5]

second_slide = X.slides.add_slide(Second_Layout)

# Agregar título para la próxima diapositiva,

second_slide.shapes.title.text = "Second slide"

#Ahora, debemos crear un cuadro de texto y mover su diseño para satisfacer nuestras necesidades.
# Posicionémoslo y ajustemos sus márgenes en pulgadas.

textbox = second_slide.shapes.add_textbox(Inches(3), Inches(1.5),Inches(3), Inches(1))

#La línea de código anterior colocará un cuadro de texto de 3 pulgadas desde la izquierda y 1.5 
#pulgadas desde la parte superior con un ancho de 3 pulgadas y una altura de 1 pulgada.

#Una vez que tengamos el diseño y la posición fijos, es hora de crear un marco de texto para agregar contenido.


textframe = textbox.text_frame

#Ahora para agregar un párrafo de contenido,


paragraph = textframe.add_paragraph()
paragraph.text = "This is a paragraph in the second slide!"

X.save("First_presentation.pptx")
```
