```python
# Crea un diccionario vacío para almacenar los contactos
contactos = {}

# Itera sobre las filas del DataFrame y agrega cada contacto al diccionario
for index, row in df.iterrows():
    # Obtiene el nombre y el número de teléfono de la fila actual
    nombre = row['Nombre completo']
    telefono = row['whatsapp']
    
    # Agrega el contacto al diccionario
    contactos[nombre] = telefono
```
