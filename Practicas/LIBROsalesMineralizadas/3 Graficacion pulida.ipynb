```python
import pandas as pd

import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

```

```python
url='https://docs.google.com/spreadsheets/d/e/2PACX-1vQMVr8Cy2DN4e5XiTfdlulCQ0daWzEP6YNmccvzgUuWre5IgJ0hmLxi-LfMpqS7T-1msiWIIDR5fehX/pub?output=csv'

df=pd.read_csv(url)
df

df= df.T.reset_index()
df

#set column names equal to values in row index position 0
df.columns = df.iloc[0]

#remove first row from DataFrame
df = df[1:]

#view updated DataFrame
df


df= df.reset_index(drop=True)
df
```

```python
df['nombreSal']
```

```python
df['supra']
```

# Matploitlib

```python
plt.bar(df['nombreSal'],df['supra'])

plt.xticks(rotation = 45)
```

```python
plt.bar(df['nombreSal'],df['Ganazufre 3-14'])
plt.xticks(rotation = 45)
```

```python
plt.bar(df['nombreSal'],df['supra'],label='Supragan')
plt.bar(df['nombreSal'],df['Ganazufre 3-14'],label='Ganazufre')
plt.legend()

plt.xticks(rotation = 45)
```

# seaborn

```python
sns.barplot(x=df['nombreSal'],y=df['supra'])

plt.xticks(rotation = 45)
```

```python
sns.histplot(data=df,x='nombreSal',y='supra')

plt.xticks(rotation = 45)
```
