```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#import data from excel 
dirSafrol='/content/dataStatistic.xlsx'

df=pd.read_excel(dirSafrol)
df
```

```python

df['conversion(%)'][1]
```

```python
#Temperatura versus porcentaje de conversion 
temperature = np.array(df['temperatureC'])
conversion = np.array(df['conversion(%)'])
four=np.array(df['four'])



plt.bar(temperature, conversion,width = 1)


plt.title('Temperatura vs conversión del safrol')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje de conversión ')

# plt.xticks(rotation='82.5')
plt.ylim(98,98.5)

plt.savefig('temperaturaVSConversion95 second.png',dpi=400)
plt.grid()
plt.show()



```

```python
plt.bar(temperature, four ,width = 1,color = "orange")

plt.title('Temperatura vs porcentaje de molecula 4')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje molecula 4 ')
plt.ylim(66,73)
plt.savefig('temperaturaVSConversionmolecula 4.png',dpi=400)
```

```python
#uso de un filtro para solo obtener datos 70 grados

idx = df['temperatureC'] > 60
idx

df_temp=df[idx]
df_temp

#ahora usao un filtro para obtener menores de 90 grados
idx = df['temperatureC'] < 90
idx

df_temp=df_temp[idx]
df_temp
```

```python
#Temperatura versus porcentaje de conversion 
temperature = np.array(df_temp['catalyst'])
conversion = np.array(df_temp['conversion(%)'])

# temperature= np.array(["A", "B", "C"])


plt.bar(temperature, conversion)


plt.title('Temperatura vs conversión del safrol')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje de conversión ')
plt.ylim(97.5,100)

# plt.xticks(rotation='82.5')

plt.savefig('temperatura70VSConversion.png',dpi=400)
# plt.grid()
plt.show()
```
