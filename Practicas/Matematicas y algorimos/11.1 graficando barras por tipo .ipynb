```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

```

```python
#import data from excel 
dirSafrol='/content/dataStatistic.xlsx'

df=pd.read_excel(dirSafrol)
df
```

```python
df['conversion(%)'][0]
```

```python
#filtro para solo  None

idx = df['catalyst'] == 'None'
idx

df_filter=df[idx]
df_filter
```

```python
#Temperatura vs conversion sin catalizador
temperature = np.array(df_filter['temperatureC'])
conversion = np.array(df_filter['conversion(%)'])




plt.bar(temperature, conversion)


plt.title('Temperatura vs conversión del Dihidrosafrol sin catalizador ')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje de conversión ')

# plt.xticks(rotation='82.5')
# plt.ylim(98,98.5)
plt.annotate("95.1 6h", (50, 98))
plt.annotate('98.1 6h', (70, 98))
plt.annotate('98.3 6h', (83, 99))
plt.annotate('32.4 14h', (10, 35))
plt.savefig('temperaturaVSConversionSinCatalizador.png',dpi=400)
# plt.grid()
plt.show()



```

```python
#filtro para todo menos sin catalizador

idx = df['catalyst'] == 'ZnCl2'
idx

df_filter=df[idx]
df_filter
```

```python
#Temperatura vs conversion sin catalizador
temperature = np.array(df_filter['temperatureC'])
conversion = np.array(df_filter['conversion(%)'])




plt.bar(temperature, conversion,width = 1)


plt.title(' conversión del Dihidrosafrol con  catalizador cloruro de zinc')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje de conversión ')

# plt.xticks(rotation='82.5')
# plt.ylim(98,98.5)
plt.annotate("95.2 4.5h", (50, 98))
plt.annotate('98 4.5h', (70, 98))
plt.annotate('41.5 14h', (10, 42))
plt.savefig('temperaturaVSConversionCatalizadorZnCl2.png',dpi=400)
# plt.grid()
plt.show()
```
