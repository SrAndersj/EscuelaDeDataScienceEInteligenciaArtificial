```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data from excel 
dirSafrol='/content/dataSafrol.xlsx'

df=pd.read_excel(dirSafrol)
df
```

```python
#Temperatura versus porcentaje de conversion 
temperature = np.array(df['relaIntensity'])
conversion = np.array(df['m/z'])



plt.bar(temperature, conversion,width = 1)

plt.title('Temperatura vs conversión del safrol')
plt.xlabel('Temperatura °C')
plt.ylabel('Porcentaje de conversión ')

# plt.xticks(rotation='82.5')

plt.savefig('temperaturaVSConversion.png',dpi=400)
plt.grid()
plt.show()
```
