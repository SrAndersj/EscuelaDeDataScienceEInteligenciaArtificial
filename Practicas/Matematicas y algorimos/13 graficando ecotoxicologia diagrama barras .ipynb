```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt


#import data from excel 
dir='/content/dataEcotoxi.xlsx'

df=pd.read_excel(dir)
df
```

```python
# type and GPx 96
t = np.array(df['type'])
g = np.array(df['GPx 96h'])




# temperature= np.array(["A", "B", "C"])

# grafica 1

plt.bar(t,g)


plt.title(' GPx 96 horas   ')
plt.ylabel('UI mg proteina de GPx ')
plt.xlabel(' Experimento')
plt.xticks(rotation=45)
# grafica 2



# plt.ylim(0.02,0.045)




plt.savefig('UI mg proteina de GPx 96 horas .png',bbox_inches = 'tight',dpi=400)   #bbox es para que guarde bien la imagen porque sin el guarda cortada la grafica 



# plt.grid()

plt.show()
```
