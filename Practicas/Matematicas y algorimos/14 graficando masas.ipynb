```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt


#import data from excel 
dir='/content/dataMasas.xlsx'

df=pd.read_excel(dir)
df
```

```python
x = np.array(df['PBOx'])
y = np.array(df['PBOy'])

plt.bar(x,y)


plt.title(' Espectro de masas PBO   ')
plt.ylabel('Intensidad Relativa')
plt.xlabel('m/z')

plt.annotate("176", (176, 9999))
plt.annotate("338", (338, 389))

plt.savefig('masasTotal .png',bbox_inches = 'tight',dpi=400)
```
