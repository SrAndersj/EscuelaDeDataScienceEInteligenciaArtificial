# given the function:
## $f(x)=x^{2}$

# its derivative is 
## $f'(x)=2x$





```python
#import

import numpy as np
import matplotlib.pyplot as plt

import sympy as sp 
```

```python
# create x-axis data

x=np.linspace(-100,100,100)


# create y-axis

y=x**2
#y
```

```python
#plotting 

fig =plt.figure()
xy=fig.subplots()
xy.plot(x,y)
```
