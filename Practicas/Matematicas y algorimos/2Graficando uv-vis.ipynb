```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.signal import savgol_filter
```

```python
dir='/content/dataSafrol.xlsx'

df=pd.read_excel(dir)
df
```

```python
longitud = df['nm']
absorvancia = df['absor']

data_plot = pd.DataFrame({"nm":longitud, "Absorvancia":absorvancia})

sns.lineplot(x = "nm", y = "Absorvancia", data=data_plot)
plt.show()
plt.savefig('foto.png')
!pwd
```

# Vamos a graficas

```python
# applying the savitsky golay filter

a=df['absor']
y_filtered =savgol_filter(a,3,1)

fig =plt.figure()
ax=fig.subplots()
#p=ax.plot(df['nm'],y_filtered,'-*')
p=ax.plot(df['nm'],y_filtered)
```

```python
longitud = df['nm']
absorvancia = y_filtered

data_plot = pd.DataFrame({"nm":longitud, "Absorvancia":absorvancia})

sns.lineplot(x = "nm", y = "Absorvancia", data=data_plot)

plt.title("Uv vis Phenol")
plt.figure(figsize=(50,50))
plt.show()
```
