```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data from excel 
dirSafrol='/content/dataSafrol.xlsx'
dirPiperonal='/content/dataPiperonal.xlsx'
dirIsosafrol='/content/dataIsosafrol.xlsx'

#create data frame d1 d2 d3
df1=pd.read_excel(dirSafrol)
df1

```

```python

df2=pd.read_excel(dirPiperonal)
df2
```

```python

df3=pd.read_excel(dirIsosafrol)
df3
```

```python
#graficando 
fig = plt.figure()
xy=fig.subplots()

#plt.figure(figsize=(4,4))
plot=xy.plot(df1['nm'],df1['absor'],label='Safrol')
plot=xy.plot(df3['nm'],df3['absor'],label="Isosafrol")
plot=xy.plot(df2['nm'],df2['absor'],label="Piperonal")

plt.legend()

xy.set_title('uv-vis Safrol y piperonal')
xy.set_xlabel('Longitud (nm)')
xy.set_ylabel('Absorbancia')





#guardando jpg
plt.savefig("safrol piperonil isosafrol.jpg", dpi=300)
```

```python

# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math
  
# Using Numpy to create an array X
X = np.arange(0, math.pi*2, 0.05)
  
# Assign variables to the y axis part of the curve
y = np.sin(X)
z = np.cos(X)
  
# Plotting both the curves simultaneously
plt.plot(X, y, color='r', label='sin')
plt.plot(X, z, color='g', label='cos')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine functions")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
