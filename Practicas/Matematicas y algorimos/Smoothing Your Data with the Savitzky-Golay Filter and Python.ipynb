```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
#generating a noisy signal
x=np.linspace(0,2*np.pi,100)
x
```

```python
y=np.sin(x)+ np.cos(x)+np.random.random(100)


#plotting 

fig =plt.figure()
ax=fig.subplots()
ax.plot(x,y)
```

```python
# applying the savitsky golay filter
y_filtered = savgol_filter(y,99,3)
```

```python
fig =plt.figure()
ax=fig.subplots()
p=ax.plot(x,y,'-*')
p=ax.plot(x,y_filtered)
```
