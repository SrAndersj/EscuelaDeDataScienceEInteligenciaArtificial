```python
import zipfile
import os

# Definimos la ruta al archivo ZIP que queremos descomprimir

zip_path = '/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/caquetaDane/18_Caqueta/18Caqueta/18_Caqueta_CSV.zip'


# Creamos una carpeta para almacenar los archivos descomprimidos
unzip_folder = '/content/unzip_folder'
if not os.path.exists(unzip_folder):
    os.makedirs(unzip_folder)

# Abrimos el archivo ZIP y lo descomprimimos en la carpeta creada anteriormente
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_folder)
    
# Imprimimos la lista de archivos descomprimidos
print(os.listdir(unzip_folder))
```

```python
import seaborn as sns

# Lee el archivo CSV
df = pd.read_csv('/content/unzip_folder/CNPV2018_5PER_A2_18.CSV')

# Crea un histograma con Seaborn
sns.histplot(data=df, x='P_EDADR', bins=range(1, 23), kde=False)

# Configura la gráfica
sns.set_style('whitegrid')
sns.despine(left=True)
plt.title('Distribución de edades')
plt.xlabel('Edad')
plt.ylabel('Número de personas')

# Muestra la gráfica
plt.show()

```
