```python
import pandas as pd

import matplotlib.pyplot as plt
from ipywidgets import interact 
```

```python
df=pd.read_excel('https://docs.google.com/spreadsheets/d/e/2PACX-1vQDq72HCeEw3kcrzwK9hTeT1ksvEX6d_QSf393_edYENgfEA6jMf8E5mv4cyAhraA/pub?output=xlsx')
df
```

```python
plt.plot(df['volumen_NaOH_ml'],df['PH'], marker='o', linestyle='-', color='b')

plt.xlabel('Volumen agregado (mL)')
plt.ylabel('pH')
plt.title('Curva de titulaci√≥n')
plt.grid(True)
plt.show()

```

```python
def plot_grafico(punto):
    plt.plot(df['volumen_NaOH_ml'][:punto], df['PH'][:punto], marker='o', linestyle='-', color='b')
    plt.xlabel('Volumen agregado (mL)')
    plt.ylabel('pH')
    plt.title('Curva de titulaci√≥n')
    plt.grid(True)
    plt.show()

interact(plot_grafico, punto=(1, len(df)))
```
