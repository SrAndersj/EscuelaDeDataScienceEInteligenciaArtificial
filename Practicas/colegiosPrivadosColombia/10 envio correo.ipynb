```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd

# cargar el DataFrame
#df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/colegiosPrivadosColombia/COLEGIOS_PRIVADOS_Y_OFICIALES.csv')

# configurar la imagen
width, height = 1920, 1080
background_color = '#FFFFFF'
font_color = '#000000'
font_size = 30
font_type = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
line_spacing = 50  # agregar 2 cm (50 pixeles) adicionales
espaciadoEnVertical = 220

# abrir la fuente
font = ImageFont.truetype(font_type, font_size)

# iterar a través de cada fila del DataFrame
for index, row in df.iterrows():

    # crear la imagen
    image = Image.open('/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/colegiosPrivadosColombia/portada.png')

    # agregar el texto
    draw = ImageDraw.Draw(image)
    nombre_colegio = row["nombreestablecimiento"]
    presentacion = "Presentación hecha para:"
    text = f"{presentacion} {nombre_colegio}"
    
    # calcular la posición del texto en la imagen
    text_width, text_height = draw.textsize(text, font=font)
    x = (width - text_width) // 2
    y = (height - text_height) // 2 + espaciadoEnVertical
    
    # escribir el texto en la imagen
    draw.text((x, y), text, font=font, fill=font_color)
    
    # agregar espacio adicional debajo del texto
    y += text_height + line_spacing

    # guardar la imagen
    image.save(f'/content/{row["nombreestablecimiento"]}.png')
```

```python
import smtplib

smtp_server = "smtp.gmail.com"
smtp_port = 587
smtp_username = ""
smtp_password = ""

smtp_connection = smtplib.SMTP(smtp_server, smtp_port)
smtp_connection.starttls()
smtp_connection.login(smtp_username, smtp_password)
```

```python
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
from email.mime.image import MIMEImage

for index, row in df.iterrows():
    establecimiento = row['nombreestablecimiento']
    correo_electronico = row['correo_Electronico']
    imagen = f"{establecimiento}.png"
    archivo_pdf = "Hoja de vida.pdf"
    carta_presentacion_pdf = "Carta de presentación profe Jhonnatan Andrés Castaño Rojas.pdf"
    codigo_pdf = "Hoja de vida.pdf"
    
    message = MIMEMultipart()
    message['From'] = smtp_username
    message['To'] = correo_electronico
    message['Subject'] = "Hoja de Vida Docente de Química"
    
    # Cuerpo del correo
    saludo = f"Hola, {establecimiento}.\n\n"
    cuerpo_correo = f"""
    
Espero quien reciba éste correo se encuentre bien.SoyJhonnatan Andres Castaño Rojas

estoy muy entusiasmado por la oportunidad de postularme como profesor de Química en
su prestigioso colegio.Permítame comenzar presentándome brevemente.Soy Químico  y 
amo nutrir y hacer crecer el Talento y
fomentar la curiosidad y amor por la ciencia.

El propósito de este correo electrónico es solicitar la oportunidad de unirme al equipo docente de {establecimiento}. 

Creo firmemente que mi experiencia y pasión por la enseñanza de la Química pueden contribuir 
significativamente al éxito de los estudiantes en su colegio.

Además, me complace compartir con ustedes que, como parte de mi enfoque en la eficiencia y la automatización,
he desarrollado una herramienta personalizada utilizando el lenguaje de programación Python.Esta herramienta
me ha permitido extraer automáticamente los nombres y correos electrónicos de colegios con el fin de agilizar
y facilitar el proceso de envío de mi hoja de vida como la presente.Al utilizar esta herramienta, he logrado 
llegar a una amplia gama de colegios de manera eficiente y efectiva y creo que la enseñanza en éste siglo debe
unir la ciencia y la tecnología.


Agradezco de antemano la oportunidad de considerar mi solicitud y estaré encantado de proporcionar más 
detalles sobre mi experiencia y mis habilidades pedagógicas en una entrevista personal.Adjunto a este 
correo encontrará mi Carta de presentacion y mi hoja de vida de Linkedin(SrAndersj),donde podrá obtener 
más información sobre mi formación académica, experiencia laboral y logros relevantes.

Aprovecho esta ocasión para expresar mi gratitud por su tiempo y consideración.Quedo a su disposición 
para responder cualquier pregunta adicional o proporcionar cualquier información adicional que puedan requerir.
Espero con entusiasmo la posibilidad de unirme al equipo de educadores de {establecimiento} y contribuir al 
crecimiento y éxito continuo de sus estudiantes.

Atentamente,

Jhonnatan Andres Castaño Rojas
321 946 50 22
jhonnatan.castano.documentos@gmail.com.


Mensaje enviado mediante codigo en python, si observa un error por favor notifiquelo en redes sociales o github SrAndersj 

Construyamos juntos el Código que transfome nuestro pais y lo haga prospero y lleno de oportunidades mediante la Ciencia y la Tecnología

"""
    cuerpo_correo = saludo + cuerpo_correo
    text = MIMEText(cuerpo_correo)
    message.attach(text)
    
    # Imagen adjunta
    with open(imagen, 'rb') as f:
        img_data = f.read()
    image = MIMEImage(img_data, name=imagen)
    message.attach(image)
    
    # Archivo PDF adjunto - Hoja de vida
    with open(archivo_pdf, 'rb') as f:
        pdf_data = f.read()
    pdf = MIMEApplication(pdf_data, name=archivo_pdf)
    message.attach(pdf)
    
    # Archivo PDF adjunto - Carta de presentacion
    with open(carta_presentacion_pdf, 'rb') as f:
        carta_data = f.read()
    carta = MIMEApplication(carta_data, name=carta_presentacion_pdf)
    message.attach(carta)

   

    
    # Envío del correo
    smtp_connection.send_message(message)
```
