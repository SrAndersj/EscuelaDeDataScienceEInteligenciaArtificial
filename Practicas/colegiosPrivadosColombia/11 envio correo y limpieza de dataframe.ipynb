# Limpiando dataframe de correos repetidos en misma columna

```python


# Función para separar correos electrónicos repetidos en diferentes filas
def separar_correos(row):
    correos = str(row['correo_Electronico'])  # Convertir a cadena de texto
    if '--' in correos:
        correos = correos.split('--')
    else:
        correos = correos.split()
    correos = [correo.strip() for correo in correos]
    return pd.Series(correos)

# Aplicar la función a cada fila del DataFrame
df = df.apply(separar_correos, axis=1).stack().reset_index(level=1, drop=True).to_frame('correo_Electronico').join(df['nombreestablecimiento'])

# Mostrar el DataFrame resultante
df
```

```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd


# configurar la imagen
width, height = 1920, 1080
background_color = '#FFFFFF'
font_color = '#000000'
font_size = 30
font_type = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
line_spacing = 50  # agregar 2 cm (50 pixeles) adicionales
espaciadoEnVertical = 220

# abrir la fuente
font = ImageFont.truetype(font_type, font_size)

# iterar a través de cada fila del DataFrame
for index, row in df.iterrows():

    # crear la imagen
    image = Image.open('/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/colegiosPrivadosColombia/portada.png')

    # agregar el texto
    draw = ImageDraw.Draw(image)
    nombre_colegio = row["nombreestablecimiento"]
    presentacion = "Presentación hecha para:"
    text = f"{presentacion} {nombre_colegio}"
    
    # calcular la posición del texto en la imagen
    text_width, text_height = draw.textsize(text, font=font)
    x = (width - text_width) // 2
    y = (height - text_height) // 2 + espaciadoEnVertical
    
    # escribir el texto en la imagen
    draw.text((x, y), text, font=font, fill=font_color)
    
    # agregar espacio adicional debajo del texto
    y += text_height + line_spacing

    # guardar la imagen
    image.save(f'/content/{row["nombreestablecimiento"]}.png')
```

# creando imagenes no duplicadas

```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd



# configurar la imagen
width, height = 1920, 1080
background_color = '#FFFFFF'
font_color = '#000000'
font_size = 30
font_type = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
line_spacing = 50  # agregar 2 cm (50 pixeles) adicionales
espaciadoEnVertical = 220

# abrir la fuente
font = ImageFont.truetype(font_type, font_size)

# crear un conjunto para almacenar los nombres de colegio procesados
nombres_procesados = set()

# iterar a través de cada fila del DataFrame
for index, row in df.iterrows():
    nombre_colegio = row["nombreestablecimiento"]
    
    # verificar si el nombre del colegio ya ha sido procesado
    if nombre_colegio in nombres_procesados:
        continue  # pasar a la siguiente fila
    
    # agregar el nombre del colegio al conjunto de nombres procesados
    nombres_procesados.add(nombre_colegio)

    # crear la imagen
    image = Image.open('/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/colegiosPrivadosColombia/portada.png')

    # agregar el texto
    draw = ImageDraw.Draw(image)
    presentacion = "Presentación hecha para:"
    text = f"{presentacion} {nombre_colegio}"
    
    # calcular la posición del texto en la imagen
    text_width, text_height = draw.textsize(text, font=font)
    x = (width - text_width) // 2
    y = (height - text_height) // 2 + espaciadoEnVertical
    
    # escribir el texto en la imagen
    draw.text((x, y), text, font=font, fill=font_color)
    
    # agregar espacio adicional debajo del texto
    y += text_height + line_spacing

    # guardar la imagen
    image.save(f'{nombre_colegio}.png')
```
