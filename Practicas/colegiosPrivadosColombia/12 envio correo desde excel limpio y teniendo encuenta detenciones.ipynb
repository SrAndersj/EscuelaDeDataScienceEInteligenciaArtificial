# creando imagenes no duplicadas

```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd



# configurar la imagen
width, height = 1920, 1080
background_color = '#FFFFFF'
font_color = '#000000'
font_size = 30
font_type = '/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf'
line_spacing = 50  # agregar 2 cm (50 pixeles) adicionales
espaciadoEnVertical = 220

# abrir la fuente
font = ImageFont.truetype(font_type, font_size)

# crear un conjunto para almacenar los nombres de colegio procesados
nombres_procesados = set()

# iterar a través de cada fila del DataFrame
for index, row in df.iterrows():
    nombre_colegio = row["nombreestablecimiento"]
    
    # verificar si el nombre del colegio ya ha sido procesado
    if nombre_colegio in nombres_procesados:
        continue  # pasar a la siguiente fila
    
    # agregar el nombre del colegio al conjunto de nombres procesados
    nombres_procesados.add(nombre_colegio)

    # crear la imagen
    image = Image.open('/content/drive/MyDrive/ColabNotebooks/EscuelaDeDataScienceEInteligenciaArtificial/Data e Inteligencia Artificial/Practicas/colegiosPrivadosColombia/portada.png')

    # agregar el texto
    draw = ImageDraw.Draw(image)
    presentacion = "Presentación hecha para:"
    text = f"{presentacion} {nombre_colegio}"
    
    # calcular la posición del texto en la imagen
    text_width, text_height = draw.textsize(text, font=font)
    x = (width - text_width) // 2
    y = (height - text_height) // 2 + espaciadoEnVertical
    
    # escribir el texto en la imagen
    draw.text((x, y), text, font=font, fill=font_color)
    
    # agregar espacio adicional debajo del texto
    y += text_height + line_spacing

    # guardar la imagen
    image.save(f'{nombre_colegio}.png')
```

```python
import smtplib

smtp_server = "smtp.gmail.com"
smtp_port = 587
smtp_username = ""
smtp_password = ""

smtp_connection = smtplib.SMTP(smtp_server, smtp_port)
smtp_connection.starttls()
smtp_connection.login(smtp_username, smtp_password)
```

```python
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
from email.mime.image import MIMEImage

contador = 0

for index, row in df.iterrows():
    establecimiento = row['nombreestablecimiento']
    correo_electronico = row['correo_Electronico']
    imagen = f"{establecimiento}.png"
    archivo_pdf = "Hoja de vida.pdf"
    carta_presentacion_pdf = "Carta de presentación profe Jhonnatan Andrés Castaño Rojas.pdf"
    codigo_pdf = "Hoja de vida.pdf"
    
    message = MIMEMultipart()
    message['From'] = smtp_username
    message['To'] = correo_electronico
    message['Subject'] = "Hoja de Vida Docente de Química"
    
    # Cuerpo del correo
    saludo = f"Hola, {establecimiento}.\n\n"
    cuerpo_correo = f"""
    ...
    # Código del cuerpo del correo
    
    ...
    
    """
    cuerpo_correo = saludo + cuerpo_correo
    text = MIMEText(cuerpo_correo)
    message.attach(text)
    
    # Adjuntar imagen
    with open(imagen, 'rb') as f:
        img_data = f.read()
    image = MIMEImage(img_data, name=imagen)
    message.attach(image)
    
    # Adjuntar archivo PDF - Hoja de vida
    with open(archivo_pdf, 'rb') as f:
        pdf_data = f.read()
    pdf = MIMEApplication(pdf_data, name=archivo_pdf)
    message.attach(pdf)
    
    # Adjuntar archivo PDF - Carta de presentación
    with open(carta_presentacion_pdf, 'rb') as f:
        carta_data = f.read()
    carta = MIMEApplication(carta_data, name=carta_presentacion_pdf)
    message.attach(carta)

    # Envío del correo
    smtp_connection.send_message(message)
    contador += 1

print(f"Se enviaron {contador} mensajes.")
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'nombreestablecimiento': ['Colegio A', 'Colegio B', 'Colegio C', 'Colegio D'],
    'correo_Electronico': ['colegioA@hotmail.com', 'colegioB@latinmail.com', 'colegioC@hotmail.com -- colegioC@latinmail.com', 'colegioD@gmail.com -- colegioD@hotmail.com']
}

df = pd.DataFrame(data)

# Función para separar correos electrónicos repetidos en diferentes filas
def separar_correos(row):
    correos = str(row['correo_Electronico'])  # Convertir a cadena de texto
    if '--' in correos:
        correos = correos.split('--')
        correos = [correo.strip() for correo in correos]
    else:
        correos = [correos.strip()]  # Si no hay '--', agregar el correo como está
    return pd.Series(correos)


# Aplicar la función a cada fila del DataFrame
df = df.apply(separar_correos, axis=1).stack().reset_index(level=1, drop=True).to_frame('correo_Electronico').join(df['nombreestablecimiento'])

# Mostrar el DataFrame resultante
print(df)
```

```python
import pandas as pd

# DataFrame con los datos de ejemplo
data = {
    'nombreestablecimiento': ['Colegio A', 'Colegio A', 'Colegio B', 'Colegio B', 'Colegio C'],
    'correo_Electronico': ['correo1@colegioa.com', 'correo2@colegioa.com', 'correo1@colegiob.com', 'correo2@colegiob.com', 'correo1@colegioc.com']
}

df = pd.DataFrame(data)

df
```

```python
# Función para combinar los correos electrónicos
def combinar_correos(group):
    return ', '.join(group)

# Aplicar el group by y combinar los correos electrónicos
df_grouped = df.groupby('nombreestablecimiento')['correo_Electronico'].apply(combinar_correos).reset_index()

# Mostrar el DataFrame resultante
df_grouped
```
