vamos a convertir al year en una variable categorica para el uso de multiples indices

```python
df_pob['year']=pd.Categorical(df_pob['year'].apply(str))
```

```python
df_pob.dtypes
```

para poner un filtro  ponemos falso para todos los paises y verdadero para los de interes

```python
idx_filtro=df_pob['Country'].isin(['Aruba','Colombia'])
idx_filtro
```

```python
df_sample=df_pob[idx_filtro]
df_sample
```

ahora vamos a crear multiples indices

```python
df_sample.set_index(['Country','year'])
```

el sort_index() es para una mejor visualizacion

```python
df_index_sample=df_sample.set_index(['Country','year']).sort_index()
df_index_sample
```
