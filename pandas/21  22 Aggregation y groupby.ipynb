# Vamos a hacer una funcion para tronsformar  Euros a dolares (asumimos que el valor de la tabla esta en euros)

````python
def mean_euro2usd(x):  
  return np.mean(x)*1.12
````

````python
mean_euro2usd([100,102])
````

# definir filtros

lo hacemos creando una funcion

la funcion trasforma euros a dolares y de eso voy a escoger la columna total factura y le digo que calcule el promedio y que me diga todoas las factuas mayores a 20



````python
def f_filter(x):
  return mean_euro2usd(x['total_bill'].mean())>20
````

````python
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
````

con el filtro anterior solo apareciron 209 elementos de los 243 , osea los que cumplen el filtro 

### aqui observamos que el filtro funcionao porque al pedir que cuente los datos  como los launch de mujeres estaban debajo de 20 entonces no pasaron al df_filtered

````python
df_filtered.groupby(['sex','time']).count()
````

# **22** Group By: extraer valor con variables categóricas


ahora vamos a extraer valor esta vez no usando las variables numericas si no tambien las mismas variables categoricas

````python
df['ones']=1
df
````

# y ahora voy a hacer grupos con estas categorias 

````python
df_g=df.groupby(['sex','smoker'])[['ones']].sum()
df_g
````

# vamos a agrupar por el nivel de los indices

que agrupe en el nivel cero es decir el genero 

````python
df_g.groupby(level=0).apply(
    
    lambda x :
    x/x.sum()*100
)
````

# otra forma de extraer valor de nuestros datos es transformar nuestras variables numericas en una variable categorica

lo vamos a hacer para convertirlo en una variable categorica para ciertos intervalos 

esto lo hacemos con cut

````python
pd.cut(df['total_bill'], bins =3)

# lo que sucede es que cada uno de los valores que estaba definido previamente
#va a quedar en una categoria 
````

# Para ver las categorias que quedaron

```
.value_counts()
```

# tambien podemos definir el ancho de nuestras categorias 


````python
pd.cut(df['total_bill'], bins =[3,18,35,60]).value_counts()
````

vamos a hacer un grupbay que tenga en cuenta el tiempo y la categoria que creamos bin_total

````python
df['bin_total']=pd.cut(df['total_bill'], bins =[3,18,35,60])
df
````

````python
df.groupby(['time','bin_total'])[['ones']].count()
````

# ahora lo anterior en porcentajes

````python
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(
    
    lambda x :
    x/x.sum()*100
)
````
