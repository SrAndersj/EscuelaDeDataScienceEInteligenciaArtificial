```python
import pandas as pd
pd.__version__
import numpy as np
```

```python
import seaborn as sns
```

```python
pd.options.display.float_format = '{:,.3f}'.format

#para restringier
# el numero de decimales que se muestra en el dataframe
```

```python
df = sns.load_dataset('tips')
df['ones']=1
df['bin_total']=pd.cut(df['total_bill'], bins =[3,18,35,60])

df.groupby(['sex','time'])[['total_bill']].mean()
```

# vamos a hacer un reset index para borrar los indices y  vamos a guardarlo den df_gp

```python
df_gp=df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

```python
df_gp.pivot_table(values='total_bill',index='sex', columns='time')
# aqui lo que veo es que mi tabla se a estructurado sex coomo  indice
```
